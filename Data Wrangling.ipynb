## Scrape lyrics

```python
import requests
try:
    from googlesearch import search
    from urllib.request import urlopen
    from bs4 import BeautifulSoup
    from bs4 import NavigableString
    from bs4 import Tag
except ImportError:
    print("No module named 'google' found")
```

```python
def get_lyrics(url):
    try:
        soup = BeautifulSoup(requests.get(url).content, 'html.parser')
        text = soup.select_one('div[class^="Lyrics__Container"], .lyrics').get_text(strip=True, separator='\n')
        #print(text)
        return text
    except AttributeError:
        return '?'
```

```python
headers_Get = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
        'Accept-Language': 'en-US,en;q=0.5',
        'Accept-Encoding': 'gzip, deflate',
        'DNT': '1',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }


def search_lyrics(artist, title):
    query = artist + " " + title + " lyrics"
    s = requests.Session()
    query = '+'.join(query.split())
    url = 'https://www.google.com/search?q=' + query + '&ie=utf-8&oe=utf-8'
    r = s.get(url, headers=headers_Get)
    #print(query)
    try:
        index_genious = r.text.index('https://genius.com/');
        index_end = r.text.find("lyrics",index_genious);
        if index_genious != -1:
            url = r.text[index_genious:(index_end + 6)]
            return get_lyrics(url)
    except ValueError:
        print("VALUE ERROR " + artist)
        return "??"
```

```python
def add_lyrics(row): 
    try:
        print(row['artist'], row['track'])
        return search_lyrics(row['artist'], row['track'])
    except OSError:
        print("OS ERROR")
        return '???'
```

```python
music_df['lyrics'] = music_df.apply(lambda row: add_lyrics(row), axis=1)
```

```python
#main

upbeat_fun_main = 'fun', 'uplifting', 'cheerful', 'gleeful', 'playful'

intense_aggressive_main = 'aggressive', 'angry','intense' 

dark_gritty_main = 'bitter', 'desperate', 'dark'

relaxed_atmospheric_main = 'relaxed', 'carefree', 'calm', 'atmospheric', 'mellow'

sad_melancholic_main = 'melancholy', 'sad', 'lonely', 'gloomy'
```

```python
def categorise_seeds_main(row):  
    for seed in row['seeds_array']:
        if seed in upbeat_fun_main:
            if (seed not in intense_aggressive_main) and (seed not in dark_gritty_main) and (seed not in relaxed_atmospheric_main) and (seed not in sad_melancholic_main):
                return 'upbeat_fun'
            else:
                return "?"
        if seed in intense_aggressive_main:
            if (seed not in upbeat_fun_main) and (seed not in dark_gritty_main) and (seed not in relaxed_atmospheric_main) and (seed not in sad_melancholic_main):
                return 'intense_aggressive'
            else:
                return "?"
        if seed in dark_gritty_main:
            if (seed not in upbeat_fun_main) and (seed not in intense_aggressive_main) and (seed not in relaxed_atmospheric_main) and (seed not in sad_melancholic_main):
                return 'dark_gritty'
            else:
                return "?"
        if seed in relaxed_atmospheric_main:
            if (seed not in upbeat_fun_main) and (seed not in intense_aggressive_main) and (seed not in dark_gritty_main) and (seed not in sad_melancholic_main):
                return 'relaxed_atmospheric'
            else:
                return "?"
        if seed in sad_melancholic_main:
            if (seed not in upbeat_fun_main) and (seed not in intense_aggressive_main) and (seed not in dark_gritty_main) and (seed not in relaxed_atmospheric_main):
                return 'sad_melancholic'
            else:
                return "?"
            
    return "?"
```

```python
music_df['sentiment'] = music_df.apply(lambda row: categorise_seeds_main(row), axis=1)
```

```python
def categorise_seeds(row):
    if row['sentiment'] != '?' and row['sentiment'] != 'multiple':
        return row['sentiment']
    uf = 0
    ia = 0
    dg = 0
    ra = 0
    sm = 0
    for seed in row['seeds_array']:
        if seed in upbeat_fun_main:
            uf+=3;
        elif seed in upbeat_fun_strong:
            uf+=2;
        elif seed in upbeat_fun:
            uf+=1;
            
        if seed in intense_aggressive_main:
            ia+=3;
        elif seed in intense_aggressive_strong:
            ia+=2;
        elif seed in intense_aggressive:
            ia+=1;
            
        if seed in dark_gritty_main:
            dg+=3;
        elif seed in dark_gritty_strong:
            dg+=2;
        elif seed in dark_gritty:
            dg+=1;
        
        if seed in relaxed_atmospheric_main:
            ra+=3;
        elif seed in relaxed_atmospheric_strong:
            ra+=2;
        elif seed in relaxed_atmospheric:
            ra+=1;
        
        if seed in sad_melancholic_main:
            sm+=3;
        elif seed in sad_melancholic_strong:
            sm+=2;
        elif seed in sad_melancholic:
            sm+=1;
    
    arr = [uf, ia, dg, ra, sm]
    max_val = max(arr)
    indices = []

    for i in range(len(arr)):
        if arr[i] == max_val:
            indices.append(i)
            
    if len(indices) > 1:
        print("multiple")
        print(indices)
        print(row['seeds_array'])
        return "multiple"
    elif len(indices) == 1:
        if indices[0] == 0:
            return "upbeat_fun"
        elif indices[0] == 1:
            return "intense_aggressive"
        elif indices[0] == 2:
            return "dark_gritty"
        elif indices[0] == 3:
            return "relaxed_atmospheric"
        elif indices[0] == 4:
            return "sad_melancholic"

    return "unknown"
```
