```python
import requests
import random
import urllib.parse
import time
from bs4 import BeautifulSoup
```

```python
# og_scrape

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate',
    'DNT': '1',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1'
}

def get_lyrics(genius_url):
    print(f'Fetching url: {genius_url}')
    try:
        soup = BeautifulSoup(requests.get(genius_url).content, 'html.parser')
        lyric_divs = soup.find_all('div', {'data-lyrics-container': True})
        parts = []
        for div in lyric_divs:
            parts.append(div.get_text(strip=True, separator='\n'))
        return '\n'.join(parts)
    except AttributeError:
        print("-------------- ATTRIBUTE ERROR --------------")
        return '?'
    
def search_lyrics_by_url(genius_url):
    try:
        return get_lyrics(genius_url)
    except ValueError:
        print("-------------- VALUE ERROR --------------")
        return "??"
    
def add_lyrics(row): 
    try:
        time.sleep(0.1 * random.randint(2, 7))
        return search_lyrics_by_url(row['genius_url'])
    except OSError:
        print("OS ERROR")
        return '???'
```

```python
split = join_df.shape[0] // 2
split
```

```python
join_df_1 = join_df.copy()[:split]
join_df_2 = join_df.copy()[split:]
join_df_1.shape, join_df_2.shape 
```

```python
join_df_1['new_lyrics'] = join_df_1.apply(lambda row: add_lyrics(row), axis=1)
```

```python
join_df_1.to_csv('processed data/music_with_lyrics_full_part1.csv')
```
